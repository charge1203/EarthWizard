```python
seoul = merged_data[merged_data['year'] == 2022]
v = seoul[['District', 'Number of Total Route', 'Length of Total Route(km)', 'year']]
x = v[v['District'] != '소계']


x.dtypes
```

```python
x['Number of Total Route'] = x['Number of Total Route'].astype('int')
x['Length of Total Route(km)'] = x['Length of Total Route(km)'].astype('float')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# 'Number of Total Route'와 'Length of Total Route(km)' 컬럼을 곱한 값을 담은 새로운 컬럼 추가
x['volume'] = x['Number of Total Route'] * x['Length of Total Route(km)']
# 구별 상대적 위치 표현
x['Latitude'] = [32, 33, 30, 40, 47, 42, 48, 36, 37, 41, 45, 25, 27, 23, 15, 10, 14, 21, 22, 26, 26, 35, 40, 49, 55]
x['Longitude'] = [40, 33, 30, 34, 34, 37, 39, 43, 47, 50, 48, 44, 37, 34, 28, 35, 20, 16, 29, 23, 15, 15, 25, 28, 33]

volume = x[['District', 'volume', 'Latitude', 'Longitude']]
# 결과 출력
volume
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 프레임에서 필요한 컬럼 추출
volume = x[['District', 'volume', 'Latitude', 'Longitude']]

# 그래프 배경색 설정
plt.rcParams['axes.facecolor'] = '#FFFCEA'

# 산점도 그래프 그리기
fig, ax = plt.subplots(figsize=(20, 15))

# 산점도 그리기
scatter = ax.scatter(volume['Latitude'], volume['Longitude'], s=volume['volume']*2, c=np.random.rand(len(volume)), alpha=0.7)

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'

# 각 산점에 District 텍스트 표시
for i, district in enumerate(volume['District']):
    ax.annotate(district, (volume['Latitude'].iloc[i], volume['Longitude'].iloc[i]), fontsize=15, ha='center')

# 컬러바 추가
plt.colorbar(scatter)

# 축 레이블 설정
# ax.set_xlabel('Latitude')
# ax.set_ylabel('Longitude')

plt.xlim([0, 60])
plt.ylim([0, 55])

# 그래프 제목 설정
plt.title('Scatter Plot')

# 그래프 표시
plt.show()
```
