```python
import pandas as pd

data = pd.read_csv("자전거도로_현황_20230713224353.csv", encoding='cp949')

data
```

```python
a = data[['시도별(1)','2021']]
a
```

```python
z = a[a['시도별(1)'] != '합계']
t = z[z['시도별(1)'] != '전국']
s = t[t['시도별(1)'] != '시도별(1)']

s = s.drop(index={10, 11, 12, 13, 14, 15, 16, 17, 18})
s
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8), facecolor='#FFFCEA')
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FF99CC', '#CC99FF', '#FFFF99', '#99FFFF']
ratio = s['2021']
labels = ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan', 'Sejong']

plt.pie(ratio, labels=labels, autopct='%.1f%%', colors=colors)
plt.title('Percentage Of Bicycle Lane by Region', fontsize=20)
plt.show()
```
