```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(40, 20))
mpl.rcParams['font.size'] = 40
from matplotlib import rc
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# 데이터프레임에서 '측정소명'과 이산화질소농도(ppm) 열을 추출
labels = df['측정소명']
values1 = df['이산화질소농도(ppm)']
values2 = df['이산화질소농도(ppm, 구별)']

# x 좌표 값
x = np.arange(len(labels))

# 막대 그래프 너비 설정
bar_width = 0.35

# 막대 그래프 그리기
plt.bar(x, values1, width=bar_width, color='blue', label='이산화질소농도(ppm)')
plt.bar(x + bar_width, values2, width=bar_width, color='red', label='이산화질소농도(ppm, 구별)')

# x 축 설정
plt.xticks(x + bar_width/2, labels, rotation=45, fontsize=30)
plt.yticks(fontsize=30)

# 축과 제목 설정
plt.xlabel('measuring stations', fontsize=40)
plt.ylabel('NO2(ppm)', fontsize=40)
plt.title('Comparison of NO2 Emissions in Districts and Roadways on 12/01/2023', fontsize=40)

# 범례 추가
plt.legend()

# 그래프 표시
plt.tight_layout()
plt.show()
```
